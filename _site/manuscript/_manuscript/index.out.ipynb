## Iris Dataset

The `iris` dataset contains measurements of sepal and petal dimensions for 150 flowers across three species: *setosa*, *versicolor*, and *virginica* ([Anderson 1935](#ref-anderson1935)).

### Sepal Dimensions

<a href="#fig-iris-sepal" class="quarto-xref">Figure 1</a> shows a clear separation between species in sepal space, with *setosa* occupying a distinct region of high width relative to length.

``` python
ggplot(iris, aes(x = Sepal.Length, y = Sepal.Width, color = Species)) +
  geom_point(alpha = 0.7, size = 2.5) +
  geom_smooth(method = "lm", se = FALSE, linewidth = 0.8) +
  scale_color_manual(values = c("#13294B", "#4B9CD3", "#99C4E2")) +
  labs(
    x = "Sepal Length (cm)",
    y = "Sepal Width (cm)",
    color = "Species"
  ) +
  theme_minimal(base_size = 13)
```

    `geom_smooth()` using formula = 'y ~ x'

<figure id="fig-iris-sepal">

<figcaption>Figure 1: Sepal length vs. width across three iris species</figcaption>
</figure>

### Petal Length

Species separation is even more pronounced in petal length (<a href="#fig-iris-petal" class="quarto-xref">Figure 2</a>), making it a reliable discriminating variable.

``` python
ggplot(iris, aes(x = Species, y = Petal.Length, fill = Species)) +
  geom_boxplot(alpha = 0.7, outlier.shape = 21) +
  geom_jitter(width = 0.15, alpha = 0.4, size = 1.5) +
  scale_fill_manual(values = c("#13294B", "#4B9CD3", "#99C4E2")) +
  labs(
    x = "Species",
    y = "Petal Length (cm)"
  ) +
  theme_minimal(base_size = 13) +
  theme(legend.position = "none")
```

<figure id="fig-iris-petal">

<figcaption>Figure 2: Distribution of petal length by species</figcaption>
</figure>

## Motor Trend Cars

The `mtcars` dataset records fuel efficiency and physical characteristics for 32 automobiles from the 1974 Motor Trend magazine ([Henderson and Velleman 1981](#ref-henderson1981)). <a href="#fig-mtcars" class="quarto-xref">Figure 3</a> illustrates the well-known negative relationship between vehicle weight and fuel efficiency, stratified by cylinder count.

``` python
mtcars |>
  mutate(cyl = factor(cyl)) |>
  ggplot(aes(x = wt, y = mpg, color = cyl)) +
  geom_point(size = 3, alpha = 0.8) +
  geom_smooth(method = "lm", se = TRUE, alpha = 0.15, linewidth = 0.9) +
  scale_color_manual(values = c("#13294B", "#4B9CD3", "#99C4E2")) +
  labs(
    x = "Weight (1000 lbs)",
    y = "Miles per Gallon",
    color = "Cylinders"
  ) +
  theme_minimal(base_size = 13)
```

    `geom_smooth()` using formula = 'y ~ x'

<figure id="fig-mtcars">

<figcaption>Figure 3: Vehicle weight vs. fuel efficiency, by number of cylinders</figcaption>
</figure>

## Conclusion

This document was rendered entirely from plain-text source files. The figures above were produced in `notebooks/analysis.qmd` and embedded without manual copying. Any change to the analysis notebook is reflected here on the next render.

## References

Anderson, Edgar. 1935. “The Irises of the Gaspe Peninsula.” *Bulletin of the American Iris Society* 59: 2–5.

Henderson, Harold V, and Paul F Velleman. 1981. “Building Multiple Regression Models Interactively.” *Biometrics* 37 (2): 391–411.
